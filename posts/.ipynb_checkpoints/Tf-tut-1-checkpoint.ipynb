````python
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
````

````python
mnist.train.images.shape
````

````python
import tensorflow as tf
````

[https://www.tensorflow.org/versions/r0.7/api_docs/python/io_ops.html#placeholder](https://www.tensorflow.org/versions/r0.7/api_docs/python/io_ops.html#placeholder)

tf.placeholder(dtype, shape=None, name=None)

Args:

* dtype: The type of elements in the tensor to be fed.
* shape: The shape of the tensor to be fed (optional). If the shape is not specified, you can feed a tensor of any shape.
* name: A name for the operation (optional).


````python
x = tf.placeholder(tf.float32, [None, 784])
````

[https://www.tensorflow.org/versions/r0.7/api_docs/python/state_ops.html#Variable](https://www.tensorflow.org/versions/r0.7/api_docs/python/state_ops.html#Variable)

The Variable() constructor requires an initial value for the variable, which can be a Tensor of any type and shape. The initial value defines the type and shape of the variable. After construction, the type and shape of the variable are fixed. The value can be changed using one of the assign methods.

More examples:
```
weights = tf.Variable(
    tf.truncated_normal([IMAGE_PIXELS, hidden1_units],
                        stddev=1.0 / math.sqrt(float(IMAGE_PIXELS))),
    name='weights')
biases = tf.Variable(tf.zeros([hidden1_units]),
                     name='biases')
```

````python
W = tf.Variable(tf.zeros([784, 10]))
b = tf.Variable(tf.zeros([10]))
````

Placeholder for external? Variable for internal?

Next: tf.nn.softmax 
[https://www.tensorflow.org/versions/r0.7/api_docs/python/nn.html#softmax](https://www.tensorflow.org/versions/r0.7/api_docs/python/nn.html#softmax)

````python
y = tf.nn.softmax(tf.matmul(x, W) + b)
````

````python
y_ = tf.placeholder(tf.float32, [None, 10])
````

[https://www.tensorflow.org/versions/r0.7/api_docs/python/math_ops.html#reduce_sum](https://www.tensorflow.org/versions/r0.7/api_docs/python/math_ops.html#reduce_sum)

````python
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
````

[https://www.tensorflow.org/versions/r0.7/api_docs/python/train.html](https://www.tensorflow.org/versions/r0.7/api_docs/python/train.html)

[https://www.tensorflow.org/versions/r0.7/api_docs/python/train.html#GradientDescentOptimizer](https://www.tensorflow.org/versions/r0.7/api_docs/python/train.html#GradientDescentOptimizer)

````python
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
````

[https://www.tensorflow.org/versions/r0.7/how_tos/variables/index.html#initialization](https://www.tensorflow.org/versions/r0.7/how_tos/variables/index.html#initialization)

````python
init = tf.initialize_all_variables() # varaibles: W and b
````

````python
sess = tf.Session()
sess.run(init)
````

````python
for i in range(1000):
  batch_xs, batch_ys = mnist.train.next_batch(100)
  # print batch_xs.shape, batch_ys.shape
  sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
````

````python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
````

````python
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
````
